# MODEL FOR GENDER

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1./255)

gender_path = "/Users/hemantpatel/Downloads/eyewear images/Gender"

train_generator = train_datagen.flow_from_directory(
        gender_path,
        target_size=(224, 224),
        batch_size=32,
        class_mode='categorical',
        shuffle=True)

num_classes = len(train_generator.class_indices)

gender_model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(num_classes, activation='softmax')
])

gender_model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

gender_model.fit(
train_generator,
steps_per_epoch=train_generator.samples//train_generator.batch_size,
epochs= 20 )

# gender_model.save('gender_model.h5')
```

# MODEL FOR SUB CATEGORY L2

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Use ImageDataGenerator to read images from directories
sub_category_path = "/Users/hemantpatel/Downloads/eyewear images/Sub Category L2"

train_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        sub_category_path,
        target_size=(150, 150),
        batch_size=32,
        class_mode='categorical')

# Define the model architecture
sub_category_model = Sequential()

sub_category_model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)))
sub_category_model.add(MaxPooling2D((2, 2)))

sub_category_model.add(Conv2D(64, (3, 3), activation='relu'))
sub_category_model.add(MaxPooling2D((2, 2)))

sub_category_model.add(Conv2D(128, (3, 3), activation='relu'))
sub_category_model.add(MaxPooling2D((2, 2)))

sub_category_model.add(Conv2D(256, (3, 3), activation='relu'))
sub_category_model.add(MaxPooling2D((2, 2)))

sub_category_model.add(Flatten())

sub_category_model.add(Dense(256, activation='relu'))
sub_category_model.add(Dropout(0.5))

num_classes = len(train_generator.class_indices)
sub_category_model.add(Dense(num_classes, activation='softmax'))

# Compile the model
sub_category_model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['acc'])

# Train the model
sub_category_model.fit(
train_generator,
steps_per_epoch=train_generator.samples//train_generator.batch_size,
epochs=20)

# Save the trained model
sub_category_model.save("sub_category_model.h5")
```

# MODEL FOR LENS COLOR

```python

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define the ImageDataGenerator for data augmentation and preprocessing
train_datagen = ImageDataGenerator(rescale=1./255,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True)

# Set the path to the directory containing the training images
lens_color_path = "/Users/hemantpatel/Downloads/eyewear images/Lens Color"

# Set the target size of the images
img_size = (224, 224)

# Set the batch size for training
batch_size = 32

# Create the training data generator
train_generator = train_datagen.flow_from_directory(lens_color_path,
                                                    target_size=img_size,
                                                    batch_size=batch_size,
                                                    class_mode='categorical')

# Get the number of classes
num_classes = len(train_generator.class_indices)

# Define the CNN model
lens_color_model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(224, 224, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(num_classes, activation='softmax')
])

# Compile the model
lens_color_model.compile(optimizer='adam',
                          loss='categorical_crossentropy',
                          metrics=['accuracy'])

# Train the model
lens_color_model.fit(
train_generator,
steps_per_epoch=train_generator.samples//train_generator.batch_size,
epochs=20)

# Save the model
lens_color_model.save('lens_color_model.h5')
```
